```python
import pandas as pd
from pathlib import Path
```

```python
bae_path = '../results_essentials/ag_news/bae3_0/summary.csv'
mango_path = '../results_essentials/ag_news/mango_0/summary.csv'
gbda_path = '../results_essentials/ag_news/gumbel_logits_adam_0/summary.csv'
```

```python
bae_df = pd.read_csv(bae_path)
bae_df
```

```python
mango_df = pd.read_csv(mango_path)
mango_df
```

```python
gbda_df = pd.read_csv(gbda_path)
gbda_df 
```

```python
def condition_mask(df):
    return (df['result_summary'] == 'OK') & (df['bert_f1'] > 0.9)

mask = condition_mask(bae_df) & condition_mask(mango_df) & condition_mask(gbda_df)
bae_df = bae_df[mask]
mango_df = mango_df[mask]
gbda_df = gbda_df[mask]
```

```python
bae_df.sample(2, random_state=0).to_csv('bae_samples.csv')
mango_df.sample(2, random_state=0).to_csv('mango_samples.csv')
gbda_df.sample(2, random_state=0).to_csv('gbda_samples.csv')
```
